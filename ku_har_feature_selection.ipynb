```python
import pandas as pd
data = pd.read_csv('/content/3.Time_domain_subsamples/KU-HAR_time_domain_subsamples_20750x300.csv', header=None)
```

```python
data.shape
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = data.values
signals = data[:, 0: 1800] 
X = np.zeros(signals.shape, dtype=np.float32)
for i in range(0,len(signals)):
    for j in range(0, 6):
        tmp = np.fft.fft(signals[i, j*300:(j+1)*300])
        X[i, j*300:(j+1)*300] = abs(tmp)
        
print(X.shape)
```

```python
y = data[:, 1800]
print(X.shape)
print(y.shape)
```

```python
!pip install skfeature-chappers
from skfeature.function.similarity_based import fisher_score
ranks = fisher_score.fisher_score(X, y) #returns the ranks of the variables based on the fisher’s score in descending order (filter method)
print(ranks)
```

```python
ranks.max()
```

```python
important_feat=ranks[0:500] # select the most 500 important features for the model 
important_feat
```

```python
X_s=pd.DataFrame(X[:, important_feat])
X_s
```

```python
y_s=pd.DataFrame(y, dtype=int)
```

```python
dataframe = pd.concat([X_s, y_s], axis=1).reindex(X_s.index)
```

```python
dataframe
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_s, y_s, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=200, max_features='sqrt', criterion='entropy')
model.fit(X_train, y_train)
acc = model.score(X_test, y_test)
acc
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("\nClassification Report")
report = classification_report(y_test, y_pred)
print(report)
# Accuracy of the model
accuracy = model.score(X_test, y_test)
print('Random forest Classifier Accuracy of the model: {:.2f}%'.format(accuracy*100))
```
